# **STEP-1 Import Library**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

# **STEP-2 Load Data from Gdrive & Check**

```
dataset = pd.read_csv('drive/My Drive/Colab Notebooks/Salary_Data.csv')
dataset.head()
```

```
# Check info & type data
dataset.info()
```

```
# Check jumlah kolom & Baris
dataset.shape
```

```
# Check Statistik 
dataset.describe()
```

# **STEP-3 Memisahkan data ke x['Masa Keja'] & y['Gaji']**

```
# memisahkan data x dan y serta merubah ke bentuk array (pandas to array)
x = dataset.loc[:,['YearsExperience']].values
y = dataset.loc[:,['Salary']].values
print(x,'\n',y)
```

# **STEP-4 Membuat Grafik Scater data asli**

```
plt.scatter(x,y)
plt.xlabel('Years_Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary',fontdict={'fontsize':20})
plt.show()
```

# **STEP-5 Membuat data Training & Testing dari data x dan y**

```
# import train_test_split dari sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

# **STEP-6 Fitting Model Regresi Linier**

```
# import regresi linier
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(x_train,y_train)
```

# **STEP-7 Visualisasi Training Linear Regressi**

```
plt.figure(figsize=(10,6))
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## **STEP-8 Visualisasi Test linear regressi**

```
# Visualising the Test set results
plt.scatter(x_test, y_test, color = 'green')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
# plt.plot(x_test, regressor.predict(x_test), color = 'green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
